### Importing Libraries

````python
from langchain_google_genai import GoogleGenerativeAI
from dotenv import load_dotenv
import os
````

````python
load_dotenv()

key = os.getenv("GOOGLE_FLASH_API_KEY") 
````

````python

llm = GoogleGenerativeAI(model="gemini-1.5-flash", api_key=key)
````

````python
user_diet:str = "I ate KFC mighty zinger burger with fries and coke"
user_height:str = "I am 5.8 feet tall"
user_weight:str = "I weigh 80 kg"
````

````python
prompt:str = f"""
Become a professional diet planner and perform following tasks:
1. Calculate the calorie count of the diet given by the user, explicitly mention each food 
item's calorie count and the quantity consumed by the user. if it is not mentioned by the user 
assume the average intake of that food item.
2. Suggest the user the best diet plan for weight loss based on its height and weight
3. Calculate the BMI of the user

'''{user_diet}'''
'''{user_height}'''
'''{user_weight}'''
"""
````

````python
result = llm.invoke(prompt)

print(result)
````
